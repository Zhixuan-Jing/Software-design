```python
import torch
from torch import nn
conv1=nn.Sequential(
      nn.Conv2d(1,4,5),
      # 20*20*4
      nn.MaxPool2d(2,2),
      #10*10*4
      nn.ReLU()
  )
conv2=nn.Sequential(
      nn.Conv2d(4,8,5),
      # 6*6*8
      nn.MaxPool2d(2,2),
      # 3*3*8
      nn.ReLU()
  )
fcon=nn.Linear(4*4*8,10)

def forward(x):
        x=conv1(x)
        x=conv2(x)
        # p4d = nn.MaxPool2d()
        x=nn.Flatten(x)
        x=fcon(x)
        return x
```
